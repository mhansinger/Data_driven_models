```python
#data_pd = pd.read_csv(' area_01.dat',delimiter='\t', header=1,names=['phi','theta','area','no_points'])

array = []

for line in open(' area_08.dat', 'r'):
    item = line.rstrip()
    array.append([float(x) for x in item.split()])
    #print(item)
```

```python
area_df= pd.DataFrame(data=array,columns=['theta','phi','area','volume','npoints'])
```
